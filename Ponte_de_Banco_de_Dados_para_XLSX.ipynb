```python
import sqlite3
import pandas as pd
```

Conectar ao banco de dados que você carregou.


```python
nome_do_arquivo_banco = "meu_primeiro_banco.db"
conexao = sqlite3.connect (nome_do_arquivo_banco)
```

2.Descobrir quais tabelas existem dentro do banco de dados.
*Isso ajuda a ter certeza do nome correto
 da tabela

```python
query_tabelas = "SELECT name FROM sqlite_master WHERE type='table'"
tabelas = pd.read_sql_query (query_tabelas,conexao)
print ("Tabelas encontradas no seu Banco de Dados:")
print (tabelas)
print("-"*50)
```

3.Puxar os dados da tabela que quer levar para o Excel.



```python
nome_tabela= "usuarios"
query_dados= f"SELECT * FROM {nome_tabela}"
df= pd.read_sql_query (query_dados,conexao)
```

4. Definindo o nome do arquivo final (.xlsx)

```python
nome_final_excel= "dados_usuarios.xlsx"
df.to_excel(nome_final_excel, index=False)
```

Fechar a conexao.

```python
conexao.close()
print(f"Pronto! Arquivo gerado com sucesso")
```
